```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import gecodb_parser
```

```python
GECODB_PATH = './COW/gecodb/gecodb_v01.tsv'
```

```python
gecodb = gecodb_parser.read(GECODB_PATH)
gecodb_filtered_10 = gecodb_parser.filter_by_freq(gecodb, min_freq=10)
gecodb_filtered_100 = gecodb_parser.filter_by_freq(gecodb, min_freq=100)
```

```python
log_freqs_orig = np.log(gecodb['freq'].values)
plt.hist(log_freqs_orig)
plt.title("GeCoDB log frequencies unfiltered"); # semicolons to suppress text output
plt.xlabel("Frequency");
plt.ylabel("Count");
```

```python
log_freqs_filtered_10 = np.log(gecodb_filtered_10['freq'].values)
plt.hist(log_freqs_filtered_10)
plt.title("GeCoDB log frequencies filtered (min=10)"); # semicolons to suppress text output
plt.xlabel("Frequency");
plt.ylabel("Count");
```

```python
log_freqs_filtered_100 = np.log(gecodb_filtered_100['freq'].values)
plt.hist(log_freqs_filtered_100)
plt.title("GeCoDB log frequencies filtered (min=100)"); # semicolons to suppress text output
plt.xlabel("Frequency");
plt.ylabel("Count");
```

```python
type_freqs_orig = gecodb_parser.type_freqs(gecodb)
plt.bar(type_freqs_orig.keys(), type_freqs_orig.values())
plt.rcParams['figure.figsize'] = [15, 5]
plt.title("GeCoDB link type frequencies unfiltered"); # semicolons to suppress text output
plt.ylabel("Count");
```

```python
type_freqs_10 = gecodb_parser.type_freqs(gecodb_filtered_10)
plt.bar(type_freqs_10.keys(), type_freqs_10.values())
plt.rcParams['figure.figsize'] = [15, 5]
plt.title("GeCoDB link type frequencies filtered (min=10)"); # semicolons to suppress text output
plt.ylabel("Count");
```

```python
type_freqs_100 = gecodb_parser.type_freqs(gecodb_filtered_100)
plt.bar(type_freqs_100.keys(), type_freqs_100.values())
plt.rcParams['figure.figsize'] = [15, 5]
plt.title("GeCoDB link type frequencies filtered (min=100)"); # semicolons to suppress text output
plt.ylabel("Count");
```
